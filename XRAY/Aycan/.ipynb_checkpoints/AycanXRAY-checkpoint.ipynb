In this project, a XRAY transformer with the given specifications below will be designed:

- Single Phase, High Frequency High Voltage Transformer
- Primary Winding Voltage ± 417 V (peak to peak 834 V for pulsing)
- Secondary Winding Voltage ± 12.5 kV (peak to peak 25 kV for pulsing)
- Rated Power 30 kW (for maximum 100 millisecond)
- Switching Frequency Minimum 100 kHz
- Ambient Temperature 0-40 °C

Firstly, material of the transformer core is selected. For this purpose, magnetisation curve of different materials are examined. 

```python
from IPython.display import Image
Image(filename='MagnetisationGraphics.png', width=550, height=550)
```

Steel has higher flux density under the same magnetic field strength compared to other materials such as iron. However, iron core has lower eddy currents. Due to this fact, iron is frequently used for high frequency applications.

I have chosen FEER type core, part number (FEER28) which can be found in this catalog: https://www.nec-tokin.com/english/product/pdf_dl/ferritecores.pdf

Dimensions given for FEER28 is as below:

```python
Acw=114.8 #window area in mm2
Acp=77 #crossectional area in mm2
```

B-H curve of the core given in the catalog is as below:

```python
from IPython.display import Image
Image(filename='BHcurve.png', width=550, height=550)
```

Now, let's define input parameters of the transformer:

```python
V_in=417 #input (primary) voltage in Volts
V_out=12500 #output (secondary) voltage in Volts
P_rated=30000 #rated output power in Watts
f=100000 #frequency in Hz
T_min=0 #min ambient temperature in Celcius
T_max=40 #max ambient temperature in Celcius
```

Operating transformer at 0.5T with 2500 A/m magnetic field intensity:

```python
B=0.5 #operating flux denisty in Tesla
H=400 #operating magnetic field intensity in A/m
e=0.90 #efficiency
```

First find primary turns Np:

```python
import math

Kf=4 #square wave coefficient
Np=(V_in)/(Kf*f*B*Acp*1e-6)#Primary number of turns using Faraday's Law
print (Np)
```

Then we can find secondary number of turns:

```python
Ns=(V_out*Np)/(V_in)
print (Ns)
```

We can approximate Np and Ns as below:

```python
Np=7
Ns=209
```

Let's find the current density from power handling capability of the core

```python
Ku=0.35 #utilization factor
J=(P_rated)/(B*f*Ku*Kf*e)
print (J)
```
