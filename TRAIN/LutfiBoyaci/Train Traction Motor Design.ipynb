### Train Traction Motor Design

-asynchronous squirrel cage induction motor-

#### Design Specifications:

Rated Power Output: 1280 kW                                    
Line-to-line voltage: 1350 V                                                                   
Number of poles: 6                                                                                   
Rated Speed: 1520 rpm (72 km/h) (driven with 78 Hz inverter)                                                
Rated Motor Torque: 7843 Nm                                                                         
Cooling: Forced Air Cooling                                                                        
Insulating Class: 200                                                                              
Train Wheel Diameter: 1210 mm                                                                                   
Maximum Speed: 140 km/h                                                                                                          
Gear Ratio: 4.821                                                                  

##### Analytical Calculations:

Followings are given:

```python
Pm = 1280000; # rated power output (W)
Vl = 1350; #line-to-line voltage
Vp = 1350/(3**(0.5)) #phase voltage
pole = 6;
pp= pole/2; #pole pair
nrated = 1520; # rated speed (rpm)
f = 78; # supply freq (Hz)
vmax = 140; # Maximum Speed (km/h)
Trated = 7843; # Rated Motor Torque (Nm)
wheel_diameter = 1.21; # (m)
gear_ratio = 4.821;
```

```python
fs= 78/pp # mechanical synchronous speed
ns=120*f/pole 
```

From the graph, for x axis value = 213, Cmech = 300 can be appropriate assumption.

```python
Cmech= 300 #kW S/m^3
x=(pi/pole)*(pp**(1.0/3.0))  #Aspect Ratio-l/D (length/ inner diameter of stator)
print 'l/Din =' ,x
Din= (Pm/1000/(Cmech*x*fs))**(1.0/3.0)
Din= round(Din*1000,0)
print 'Din, inner diameter=', Din , 'mm'
l=round(Din*x)
print 'l, axial length=', l , 'mm'
```

```python
Do= round (1.78*Din)
print 'Outer diameter =', Do ,'mm'
```

```python
g=round(1.6*(0.18+0.006*Pm**0.4),2) # %60 safety factor is added
print 'Air gap =', g ,'mm'
```

```python
from IPython.display import Image 
Image(filename='fluxdensity.png')
```

assume: Flux density = 0.75 T in airgap

```python
Ftan = Trated/(Din/2/1000) # N
print 'Ftan =', Ftan ,'N'
surface_area = pi*(Din/1000)*(l/1000)  #m^2
print 'surface_area =', surface_area ,'m^2'
sheer_stress = Ftan/surface_area   #N/m^2=Pa
print 'sheer_stress =', sheer_stress ,'N/m'
Bg = 0.75 # magnetic loading is assigned as 0.75 T
print 'magnetic loading, Bg =', Bg ,'T'
electric_loading = sheer_stress/Bg/1000; # kA/m
print 'electric_loading =', electric_loading ,'kA/m'
```

Motor has 6 poles and 3 phases. Therefore, slot number should be product of 6*3=18. Also, in induction machines and PMSMs, stator slot pitch can be chosen between 7-45 mm. 

```python
Stator_inner_cicumference= pi*Din
print 'Stator_inner_cicumference' , Stator_inner_cicumference , 'mm'
Minimum_slot_number= round(Stator_inner_cicumference/45)
print 'Minimum_slot_number' , Minimum_slot_number 
Maximum_slot_number= round(Stator_inner_cicumference/7)
print 'Maximum_slot_number' , Maximum_slot_number
```

```python
from IPython.display import Image 
Image(filename='slotnumber.png')
```

Above figure shows us the most advantageous slot numbers. Choosing low number of slots could lead to Increased leakage inductance , reduced breakdown torque and larger MMF harmonics. let's assign Qs=72

```python
Qs=72 #stator slot number
q=Qs/pole/3 # slots per pole per phase
print 'slots per pole per phase=' , q
```

1 phase is distributed to 4 slots

```python
alpha= pi/q/3 #slot angle
print 'stator slot angle=' , alpha , 'radian'
slot_pitch= Stator_inner_cicumference/Qs #in mm
print 'slot_pitch=' , slot_pitch , 'mm'
```

It is assigned that two layer winding with chorded coils 5/6.

```python
kp=sin(pi/2*5/6) #chording factor
print 'kp (chording factor)=' , kp
kd=(sin(q*alpha/2))/(q*sin(alpha/2)) #zone factor
print 'kd (zone factor)    =' , kd 
kw=kd*kp #winding factor
print 'kw (winding factor) =' , kw 
```

```python
#from the book, the eqn 6.9: flux=l*polepitch*Bav
polepitch=pi*Din/pole
print 'pole pitch   =' , polepitch, 'mm'
Bav=Bg*2/pi
print 'Bav   =' , Bav, 'T'
Flux_per_pole=l/1000*(polepitch/1000)*Bav
print 'Flux per pole  =' , Flux_per_pole, 'Weber'
```

```python
#remember: electric_loading = 40.5972183709 A/m
#electric loading = linear current density = slot current / slot pitch  , eqn-6.10
#Is=electric_loading*slot_pitch/1000 # slot current
#print 'Is (slot current)=', Is, 'A' BU KISIMDAN EMİN DEĞİLİM
#Erms=1/(2**0.5)*2*pi*f*kw*N*Flux ---- eqn6.6
Erms=Vp/sqrt(2)
print 'Erms =' , Erms, 'Vrms'
N=Erms*sqrt(2)/2/pi/f/kw/Flux_per_pole
print 'N (number of winding turns in a phase) =', N
```

Let's assign N=24 which is a product of slots per pole per phase and pole pair.                           
Bg value will also be recounted with the new turn ratio value

```python
N=24
print 'N, number of winding turns in a phase =' , N
a= 1 #number of current paths in parallel
z= a*N/(pp*q) # number of conductors in the slot 
print 'z, number of conductors in the slot  =' , z
Bgnew=25.2/N*Bg
print 'Bgnew', Bgnew, 'T'
```

```python
efficiency= 0.95 #efficiency is assigned
Pin=Pm/efficiency # input power 
print  'Input power', Pin, 'W'
pf=0.85 #power factor
Ip=Pin/3/Vp/pf
print 'Phase current', Ip , 'A'
J=6.5 #A/mm^2 current density is assigned
copper_area= Ip/J
print 'needed copper area', copper_area , 'mm^2'
```

skin depth:

```python
p=1.68e-08*(1+0.0039*(100-20)) #resistivity for copper @ 100*C
w=2*pi*f
u=4e-07*pi*0.999991 # permeability of air*relative mag. permeability of copper
SD=sqrt(2*p/(w*u))*1000 #skin depth in mm
print 'Skin Depth =', SD ,'mm'
```

4 AWG3 cables can be bundled. It's area is 26.7 mm^2 and radius is 2.915 mm. Skin depth is greater than radius and there is no skin effect problem. 

###### Stator Slot Sizing:

```python
#useful slot area
awg_area=4*26.7
kfill=0.4
Asu=awg_area*z/kfill
print 'Useful Slot Area=' ,Asu, 'mm^2'
```

```python
from IPython.display import Image 
Image(filename='statorlot.png')
```

```python
Bts=1.6  #tooth flux density is assigned
Kfe= 0.965 #space factor of iron
bts=Bg*slot_pitch/Bts/Kfe
print 'stator tooth width', bts, 'mm'
```

```python
#assign:
bos = 3 
hos = 1.5
hw = 2
bs1=((pi*(Din+(2*hos)+(2*hw)))/Qs)-bts
print 'bs1=', bs1 , 'mm'
bs2=sqrt(4*Asu*tan(pi/Qs)+bs1**2)
print 'bs2=' , bs2, 'mm'
hs=(2*Asu)/(bs1+bs2)
print 'hs=' ,hs, 'mm'
hcs=(Do/2)-(Din/2)-(hos+hw+hs)
print 'hcs=' ,hcs, 'mm'
```

```python
Bcs=(((pi/2)*Flux_per_pole)/(2*l*1e-3*hcs*1e-3)) #back core flux density
print 'Bcs=',Bcs,'T'
```

0.6 is not between the recommended back core flux density=1.4-1.7T. Let's decrease the Stator outer diameter.

```python
#assign 
Do=845 # mm 
hcs=(Do/2)-(Din/2)-(hos+hw+hs)
print 'hcs=' ,hcs, 'mm'
Bcs=((pi/2)*Flux_per_pole/(2*l*1e-3*hcs*1e-3)) #back core flux density
print 'Bcs=',Bcs,'T'
```

###### Rotor Slot Sizing:

```python
KI= 0.8*pf+0.2
Ib=KI*2*3*N*kw*Ip/Qr #bar current
print 'bar current,Ib=', Ib ,'A'
Jb=4.5 #A/mm^2 rotor bar current density
Ab= Ib/Jb #rotor slot area
print 'rotor slot area, Ab=', Ab ,'mm^2'
Ier = Ib/(2*sin(pp*pi/Qr)); # A
print 'end ring current ,Ier=', Ier ,'A'
Jer = 0.75*Jb; # A/mm^2 end ring current density
Aer = Ier/Jer #end ring cross section
print 'end ring cross section, Aer=', Aer ,'mm^2'
```

```python
rotor_slot_pitch= (Din-g)*pi/Qr
print 'rotor_slot_pitch=', rotor_slot_pitch, 'mm'
Btr= 1.6 #T #rotor tooth flux density is assigned
btr=Bg*rotor_slot_pitch/Btr/Kfe
print 'rotor tooth width, btr=', btr, 'mm'
```

```python
from IPython.display import Image 
Image(filename='rotorslot.png')
```

```python
bor=2.5 #mm
print 'bor=', bor, 'mm'
hor=1 #mm
print 'hor=', hor, 'mm'
Dro=Din-2*g
print 'Dro=', Dro, 'mm'
a1=((pi*(Dro-2*hos))/Qr)-btr
print 'a1=', a1, 'mm'
a2=sqrt(4*Ab*tan(pi/Qr)+a1**2)
print'a2=', a2, 'mm'
b= (2*Ab)/(a1+a2)
print 'b=', b, 'mm'
```

```python
Bcr=1.6 # rotor back core rotor flux density
hcr=((Flux_per_pole)/(2*l*1e-3*Bcr))*1000 #back core length
print 'back core length ,hcr=',hcr,'mm'

Dshaft=Dro-2*(hor+b+hcr)
print 'Dshaft=', Dshaft,'mm'
```

above formulas are going to be used

```python
kc1=slot_pitch/(slot_pitch-(bos**2)/(5*g+bos))
print 'kc1=', kc1
kc2=rotor_slot_pitch/(rotor_slot_pitch-(bor**2)/(5*g+bor))
print 'kc2=', kc2
kc=kc1*kc2
print 'Carters coeff, kc=', kc
```

```python
from IPython.display import Image 
Image(filename='magnetization.png')
```

```python
from IPython.display import Image 
Image(filename='mmf.png')
```

```python
Fmg=((kc*g/1000*Bgnew)/(4*pi*1e-7))
print 'Airgap mmf, Fmg', Fmg, 'Aturns'
#stator tooth flux density, Bts=1.6 T therefore H=2460 A/m
Hts=2460
Fmts=Hts*(hs+hw+hos)/1000 #stator tooth MMF
print'Stator tooth MMF, Fmts=', Fmts,'Aturns'
#stator tooth flux density, Btr=1.6 T therefore H=2460 A/m
Fmtr=Hts*(b+hor)/1000 #rotor tooth MMF
print'Rotor tooth MFF, Fmtr=', Fmtr,'Aturns'
```

```python
Ccs=0.88*exp(-0.4*(Bcs**2)) #remember Bcs=1.42
Ccr=0.88*exp(-0.4*(Bcr**2)) #remember Bcr=1.6
Hcs=900 # A/m 
Fmcs=Ccs*(pi*(Do-hcs)*Hcs)/(2*pp)/1000 #Stator back core MMF
Hcr=2460 #A/m
Fmcr=Ccr*(pi*(Dshaft+hcr)*Hcr/(2*pp))/1000 #Rotor back core MMF
print 'Stator back core MMF, Fmcs =',Fmcs,'Aturns'
print 'Rotor back core MMF, Fmcr =',Fmcr,'Aturns'
```

```python
#Magnetizing MMF
F1m=2*(Fmg+Fmts+Fmtr+Fmcs+Fmcr) 
print 'Magnetizing MMF, F1m=',F1m,'Aturns'
```

```python
#magnetization current
Iu=(pi*pp*F1m)/(2*3*sqrt(2)*N*kw) 
print 'Magnetizzation current, Iu=',Iu,'A'
Iu_pu=Iu/Ip
print 'Iu_pu=',Iu_pu,'pu'
```

```python
y=slot_pitch*5/6 #coil span
print 'y=',y,'mm'
l_end=(pi/2)*y+18 #end connection length
print 'l_end=', l_end,'mm'
l_coil=2*(l+l_end) # coil length
print 'l_coil=',l_coil,'mm'
Pco=1.78*1e-8 *80/273+1.78*1e-8 #copper resistivity at 100 C 
Rs=(Pco*l_coil*N)*1000/(awg_area) #Stator resistance
print 'Rs=',Rs,'Ohm'
```

```python
l_er=(pi*(Der-b_end_ring))/Qr #end ring segment length 
print 'l_er=',l_er,'mm'
Rbe=(Pal_80*l*KR/Ab+(Pal_80*l_er)/(2*Aer*sin(pi*pp/Qr)*sin(pi*pp/Qr)))*1000  
print 'Rbe=',Rbe,'Ohm'
```
