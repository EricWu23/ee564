### Train Traction Motor Design

-asynchronous squirrel cage induction motor-

#### Design Specifications:

Rated Power Output: 1280 kW                                    
Line-to-line voltage: 1350 V                                                                   
Number of poles: 6                                                                                   
Rated Speed: 1520 rpm (72 km/h) (driven with 78 Hz inverter)                                                
Rated Motor Torque: 7843 Nm                                                                         
Cooling: Forced Air Cooling                                                                        
Insulating Class: 200                                                                              
Train Wheel Diameter: 1210 mm                                                                                   
Maximum Speed: 140 km/h                                                                                                          
Gear Ratio: 4.821                                                                  

##### Analytical Calculations:

Followings are given:

```python
Pm = 1280000; # rated power output (W)
Vl = 1350; #line-to-line voltage
Vp = 1350/(3**(0.5)) #phase voltage
pole = 6;
pp= pole/2; #pole pair
nrated = 1520; # rated speed (rpm)
f = 78; # supply freq (Hz)
vmax = 140; # Maximum Speed (km/h)
Trated = 7843; # Rated Motor Torque (Nm)
wheel_diameter = 1.21; # (m)
gear_ratio = 4.821;
```

```python
fs= 78/pp # mechanical synchronous speed
ns=120*f/pole 
```

From the graph, for x axis value = 213, Cmech = 300 can be appropriate assumption.

```python
Cmech= 300 #kW S/m^3
x=(pi/pole)*(pp**(1.0/3.0))  #Aspect Ratio-l/D (length/ inner diameter of stator)
print 'l/Din =' ,x
Din= (Pm/1000/(Cmech*x*fs))**(1.0/3.0)
Din= round(Din*1000,0)
print 'Din, inner diameter=', Din , 'mm'
l=round(Din*x)
print 'l, axial length=', l , 'mm'
```

```python
g=round(1.6*(0.18+0.006*Pm**0.4),2) # %60 safety factor is added
print 'Air gap =', g ,'mm'
```

assume: Flux density = 0.75 T in airgap

```python
Ftan = Trated/(Din/2/1000) # N
print 'Ftan =', Ftan ,'N'
surface_area = pi*(Din/1000)*(l/1000)  #m^2
print 'surface_area =', surface_area ,'m^2'
sheer_stress = Ftan/surface_area   #N/m^2=Pa
print 'sheer_stress =', sheer_stress ,'N/m'
Bg = 0.75 # magnetic loading is assigned as 0.75 T
print 'magnetic loading, Bg =', Bg ,'T'
electric_loading = sheer_stress/Bg/1000; # kA/m
print 'electric_loading =', electric_loading ,'kA/m'
```

Motor has 6 poles and 3 phases. Therefore, slot number should be product of 6*3=18. Also, in induction machines and PMSMs, stator slot pitch can be chosen between 7-45 mm. 

```python
Stator_inner_cicumference= pi*Din
print 'Stator_inner_cicumference' , Stator_inner_cicumference , 'mm'
Minimum_slot_number= round(Stator_inner_cicumference/45)
print 'Minimum_slot_number' , Minimum_slot_number 
Maximum_slot_number= round(Stator_inner_cicumference/7)
print 'Maximum_slot_number' , Maximum_slot_number
```

```python
from IPython.display import Image 
Image(filename='slotnumber.png')
```

Above figure shows us the most advantageous slot numbers. Choosing low number of slots could lead to Increased leakage inductance , reduced breakdown torque and larger MMF harmonics. let's assign Qs=72

```python
Qs=72 #stator slot number
q=Qs/pole/3 # slots per pole per phase
print 'slots per pole per phase=' , q
```

1 phase is distributed to 4 slots

```python
alpha= pi/q/3 #slot angle
print 'stator slot angle=' , alpha , 'radian'
slot_pitch= Stator_inner_cicumference/Qs #in mm
print 'slot_pitch=' , slot_pitch , 'mm'
```

It is assigned that two layer winding with chorded coils 5/6.

```python
kp=sin(pi/2*5/6) #chording factor
print 'kp (chording factor)=' , kp
kd=(sin(q*alpha/2))/(q*sin(alpha/2)) #zone factor
print 'kd (zone factor)    =' , kd 
kw=kd*kp #winding factor
print 'kw (winding factor) =' , kw 
```

```python
#from the book, the eqn 6.9: flux=l*polepitch*Bav
polepitch=pi*Din/pole
print 'pole pitch   =' , polepitch, 'mm'
Bav=Bg*2/pi
print 'Bav   =' , Bav, 'T'
Flux_per_pole=l/1000*(polepitch/1000)*Bav
print 'Flux per pole  =' , Flux_per_pole, 'Weber'
```

```python
#remember: electric_loading = 40.5972183709 A/m
#electric loading = linear current density = slot current / slot pitch  , eqn-6.10
#Is=electric_loading*slot_pitch/1000 # slot current
#print 'Is (slot current)=', Is, 'A' BU KISIMDAN EMİN DEĞİLİM
#Erms=1/(2**0.5)*2*pi*f*kw*N*Flux ---- eqn6.6
Erms=Vp/sqrt(2)
print 'Erms =' , Erms, 'Vrms'
N=Erms*sqrt(2)/2/pi/f/kw/Flux_per_pole
print 'N (number of winding turns in a phase) =', N
```

Let's assign N=24 which is a product of slots per pole per phase and pole pair.

```python
N=24
print 'N, number of winding turns in a phase =' , N
a= 1 #number of current paths in parallel
z= a*N/(pp*q) # number of conductors in the slot 
print 'z, number of conductors in the slot  =' , z
Bgnew=25.2/N*Bg
print 'Bgnew', Bgnew, 'T'
```

skin depth:

```python
p=1.68e-08*(1+0.0039*(100-20)) #resistivity for copper @ 100*C
w=2*pi*f
u=4e-07*pi*0.999991 # permeability of air*relative mag. permeability of copper
SD=sqrt(2*p/(w*u))*1000 #skin depth in mm
print 'Skin Depth =', SD ,'mm'
```
