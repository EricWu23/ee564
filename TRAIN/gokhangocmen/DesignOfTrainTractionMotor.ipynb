## Step 1) Defining machine constant "Cmech"

Cmech is a machine constant of totally enclosed asynchronous and synchronous machines as a function of pole power. ## Step 3) Determining the airgap length

In machines in which the magnetizing current is taken from the supply network, the length of the air gap is dimensioned to produce a minimum magnetizing current and, on the other hand, an optimal efficiency. A small air gap gives a low magnetizing current. A small air gap also increases the surface losses in the rotor caused by the current linkage harmonics of the stator. There is no theoretical optimum solution to define airgap length but empirical solutions may be employed as given below formula which is given for 50Hz operation.In our application, it is driven by 78Hz inverter fed.However,below formula gives an approach to find the airgap length.

```python
from IPython.display import Image 
Image(filename='airgap.jpg')
```

```python
airgap=(0.18+0.006*Pout**(0.4))/1000; #given formula for pole pairs higher than 1.
print 'Airgap length= ',airgap,'in units of meter'
airgap_mm=airgap*1000;
print 'Airgap length= ',airgap_mm,'in units of milimeter'
```

In frequency converter drives, the air gap may be increased similarly as in heavy-duty
drives (60% increase) to get lower rotor surface losses.

```python
airgap_mm_rel=airgap_mm*1.6;
print 'Realiable airgap length= ',airgap_mm_rel,'in units of milimeter'
design_airgap_mm_opt=round(airgap_mm_rel);
print 'Optimized airgap length for design = ',design_airgap_mm_opt,'in units of milimeter'
design_airgap_m_opt=design_airgap_mm_opt/1000;
```

## Step 5) Finding the number of slots and slot pitches of stator

A guiding principle is that a poly-phase winding produces the more sinusoidal current linkage, the more slots there are in the stator.The winding factors for an integer slot winding usually become lower for the harmonics as the number of slots increases (the current linkage produced by the winding approaches sinusoidal waveform). Now, in principle, the current linkage created by the stator currents has a smooth distribution and, because of a long air gap, an optimum sinusoidal flux density distribution on the rotor surface. A large number of slots increases the number of coils and also the price of the machine. The lowest slot pitches occur in small machines, whereas the highest slot pitches are found in large machines. Here below, the recommendend slot pitches shown for different types of machines. Integral slot windings are recommended for squirrel cage motors.

```python
from IPython.display import Image 
Image(filename='slotscriteria.jpg')
```

```python
#Check for 3 slots/pole/phase
qs_3=3; #slots/pole/phase
Qs_3spp=2*pp*m*qs_3; #slot numbers
Tus_3spp=(pi*Di_mm)/Qs_3spp;#slot pitch in units of mm
print 'Slot pitch= ',Tus_3spp,'in units of mm for 3 slots/pole/phase'
#Check for 4 slots/pole/phase
qs_4=4; #slots/pole/phase
Qs_4spp=2*pp*m*qs_4; #slot numbers
Tus_4spp=(pi*Di_mm)/Qs_4spp;#slot pitch in units of mm
print 'Slot pitch= ',Tus_4spp,'in units of mm for 4 slots/pole/phase'
#Check for 5 slots/pole/phase
qs_5=5; #slots/pole/phase
Qs_5spp=2*pp*m*qs_5; #slot numbers
Tus_5spp=(pi*Di_mm)/Qs_5spp;#slot pitch in units of mm
print 'Slot pitch= ',Tus_5spp,'in units of mm for 5 slots/pole/phase'
```

## Step 6) Finding the airgap flux density and number of turns/phase

According to the table below, we see that airgap flux denstiy for asynchronous machines is in the interval of 0.7T and 0.9T.We can choose the airgap flux density as 0.8T, but we should check linear current density for Table 6.4 for compliance.

```python
from IPython.display import Image 
Image(filename='airgapfluxdenst.jpg')
```

According to Table 6.2 the air-gap flux density varies normally from 0.7 T to 0.9 T. It is chosen as 0.8T.

```python
Bgap=0.8; #airgap peak flux density is chosen as 0.8T 
ksi=0.8;
#Cosine of the angle ζ between the linear current density and the air-gap flux density is assumed to be 0.8
As=(ftangential*sqrt(2))/(Bgap*ksi);
print 'Linear current density= ',As,'in units of A/m'
print 'Airgap flux density= ',Bgap,'in units of T'
```

The linear current density is in the acceptable range for according to table 6.4 which has a range from 30000-65000 A/m

As the main dimensions, the winding method and the air-gap density have been selected, the
required number of winding turns N can be defined with a desired emf.The emf Em = ω.Ψm
induced by the air-gap flux linkage (Ψm = Im.Lm) can first be estimated from the RMS value
Us (Vl) of the fundamental terminal voltage; for induction motors it is Em ≈ (0.93 − 0.98)Us

```python
Us=Vl;
Em=0.97*Us;
w=2*pi*fe; #assuming small saturation
ai=2/(pi);#coefficient showing the arithmetical average of the flux density of one pole
#The more the teeth of the machine are saturated, the higher the value of αi becomes.
Tp=(pi*Di)/(2*pp); #Di is the stator inner diameter
kps1=sin(g*(pi/2));
kds1=sin(pi/(2*m))/(Qs_4spp*sin(pi/(2*m*Qs_4spp)))
kw1=kps1*kds1;
print 'Windin factor for 1st harmonic',kw1,'.'
Ns_try=(sqrt(2)*Em)/(w*kw1*l*Tp*ai*Bgap)
Ns=round(Ns_try)
print 'The number of the coil turns in series in a phase winding= ',Ns,'in units of turns'
```

If possible, it is often advisable to connect the coils in series, since in that case the possible
asymmetries between the pole pairs do not cause circulating currents in the machine.However,
parallel paths result in a larger number of slot conductors zQs which is favorable when trying
to avoid skin effect and circulating currents in the windings. 

```python
a=1; # all conductors are series no parallel paths
zQ_try=(2*a*m*Ns)/Qs_4spp
zQ=round(zQ_try);
print 'The number of conductors per slot= ',zQ,'.'
```

```python
Bgap_new=(sqrt(2)*Em)/(w*kw1*l*Tp*ai*Ns);
print 'The new air-gap flux density= ',Bgap_new,'in units of T.'
```
