## Step 1) Defining machine constant "Cmech"

The parameter defined in the literature as machine constant, Cmech, will be used practically. Cmech is a machine constant of totally enclosed asynchronous and synchronous machines as a function of pole power. Magnetic and electrical loading cannot be used as a determination because of unknown parameters slot pitch,diameter and length of the machines.Highly effective cooling methods increase the permitted electromagnetic loading and correspondingly the weight of the machine drops by 30–50 %.

```python
Pmech=Pout_kw; #in units of kw of rated output power of motor
#D is the diameter of the airgap
#p is the pole pair in the book of Jha pyrhönen
#x axis of the below curve is Pmech/(2*pp)
Pmech_curve=Pmech/(2*pp);
print 'Pole Power = ',Pmech_curve,'in units of kW'
```

```python
Cmech=275; #found from the graph which is in units of kWs/m3
print 'Machine constant = ',Cmech,'kWs/m3'
```

Pmech=Cmech*(D^^2*nsync*l), where D is the airgap diameter, l is the machine length, nsync is the synchronous speed. By defining synchronous speed, D^^2 x l product will be found.

## Step 2) Finding synchronous speed and dimensions of machine

## Step 3) Determining the airgap length

In machines in which the magnetizing current is taken from the supply network, the length of the air gap is dimensioned to produce a minimum magnetizing current and, on the other hand, an optimal efficiency. A small air gap gives a low magnetizing current. A small air gap also increases the surface losses in the rotor caused by the current linkage harmonics of the stator. There is no theoretical optimum solution to define airgap length but empirical solutions may be employed as given below formula which is given for 50Hz operation.In our application, it is driven by 78Hz inverter fed.However,below formula gives an approach to find the airgap length.

```python
from IPython.display import Image 
Image(filename='airgap.jpg')
```

```python
airgap=(0.18+0.006*Pout**(0.4))/1000; #given formula for pole pairs higher than 1.
print 'Airgap length= ',airgap,'in units of meter'
airgap_mm=airgap*1000;
print 'Airgap length= ',airgap_mm,'in units of milimeter'
```

In frequency converter drives, the air gap may be increased similarly as in heavy-duty
drives (60% increase) to get lower rotor surface losses.

```python
airgap_mm_rel=airgap_mm*1.6;
print 'Realiable airgap length= ',airgap_mm_rel,'in units of milimeter'
design_airgap_mm_opt=round(airgap_mm_rel);
print 'Optimized airgap length for design = ',design_airgap_mm_opt,'in units of milimeter'
design_airgap_m_opt=design_airgap_mm_opt/1000;
```

## Step 4) Determining the stator core length,stator inner diameter and rotor outer diameter

As it was found above, the average airgap diameter D, length of motor (l=l') and airgap length. We can calculate stator core length, by assuming no cooling channels in the stator.

```python
ls=l-2*design_airgap_m_opt;
print 'Stator core length= ',ls,'in units of meter'
```
