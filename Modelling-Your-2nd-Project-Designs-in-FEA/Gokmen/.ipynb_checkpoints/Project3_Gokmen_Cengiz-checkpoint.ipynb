At first look, we can say that theoretical design parameters are suitable with the results of RMxprt tool. The power output of the generator is 249.6kW with an efficiency of 91% and at a power factor of 0.92. At project 2, we have aimed 90% efficiency and 0.85 power factor. Thus, the results are appropriate.

Another data is rated torque and phase current values, which are 3226 Nm and 412A. We have found at Project 2 these values as 3248Nm and 417A respectively. Hence, we can also say that these values are suitable with each other.

Now, we can examine the metric data whether this design is applicable to produce or not. Thus, we will look two parameters: saturation and fill factor. 

If we look at the section of "Detailed Data at Rated Operation", we can see that all flux densities are below 1.8T, which is the absolute limit for steel laminations. Thus, we can say that generator can operate at rated speed and torque without no saturation. However, stator-teeth flux density, which is 1.73T, is really close to the limit. Thus, this constraint will be examined at Maxwell 2D again.

Stator slot fill factor at the metrics can be seen as warning saying that this design is infeasible since fill factor is beyond 75%, which is calculated as 154%!. Thus, we can express that this design needs a change. We should increase the volume of stator slots. However, we also know that stator-teeth flux density is at the limit. Hence, to decide how we will change the stator slots, we need a Maxwell 2D design (The other design issues will be discussed at the 2nd iteration of the design).

## Maxwell 2D Design 

Importing RMxprt design to Maxwell 2D, we have started to a transient simulation of the motor. At this simulation, shaft of the motor were rotating with a speed of 740rpm at no-load. We have also considered eddy effect at both stator coils and rotor bars. A snapshot of the flux density of whole motor has been received at below. As seen, there are some points such that flux density reaches to 3.16T! This is an infeasible design. To see better where the saturation occurs more, we have scaled flux density bar to 1.8T and observed the next picture.

```python
Image(filename='flux_density_1st_1.png', width=700, height=700)
```

```python
Image(filename='flux_density_1st_2.png', width=700, height=700)
```

As seen at the upper picture, stator back core has high saturation rate. If we closely look at stator teeth at below, we can also see that stator teeth also saturated.

```python
Image(filename='flux_density_1st_3.png', width=700, height=700)
```

In another evaluation can be done by looking flux lines at below. It can be seen that flux lines go across the stator slots instead of passing from teeth. This results in leakage flux and increased leakage inductance.

```python
Image(filename='flux_lines_1st_1.png', width=700, height=700)
```

To sum up, we need to increase diameter of stator to prevent saturation at the back core. Another change is we need to widen stator teeth to decrease saturation, too. Since we have also seen that slot fill factor is beyond the limit, we should decrease the width of the stator and increase the height of it. By doing the expressed changes, we have implemented 2nd iteration at RMxprt.

## Modelling in RMxprt - 2nd iteration

At this iteration, we have made those changes:
 - outer diameter has been increased from 590mm to 630mm
 - stator teeth width has been increased from 5.2mm to 7.2mm
 - stator slot height has been increased from 14.3mm to 34.3mm 
 - iron core loss components have been added to BH curve of the steel

## Conclusion

In this project, we have evaluated our design conducted at Project 2 with Maxwell simualtion program. We have got the chance of validating our generator and to see the lacks/deficiencies.

The first lack was generator has regions that saturate at some moments of rotation. This is solved by increasing stator outer diameter and increasing height of stator slots. In addition, there were saturation points between teeth and this is solved by increasing stator teeth width 2mm. 

The second lack was the slot fill factor. This is achieved by increasing the volume of the stator slots.

After correcting the values, there have been comparisons between FEA and 2nd project from the perspective of operating characteristics, losses and efficiency, main dimensions and operating values, phase currents, output power and torque. The results show that FEA and 2nd project results close each other although there are some variations at breakdown torque and resistances.

Maxwell FEA tool is efficient to see flux densities and flux lines through all the regions of generator. It gives the flexibility of both examining at magnetostatic analysis and transient analysis. While magnetostatic analysis reveals the magnetic characteristics of the material, transient analysis give the chance of rotating shaft analysis and applied torque operation by sweeping time. Doing that, designer can see many factors efficiently without producing the motor.
