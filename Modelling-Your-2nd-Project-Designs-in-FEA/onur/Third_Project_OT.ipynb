## 3. Modelling

To model the induction motor, firstly three phase induction motor parameters have been entered into RMxprt; then, stator and rotor dimensions of the motor have been obtained. Then, with simulating the motor in RMxprt, some motor characteristics have been obtained such as torque vs. slip, torque vs. speed, output power vs. speed, efficiency vs. speed. Next, the model which was created in RMxprt has been imported to the Maxwell 2D. In  Maxwell 2D finite element analysis software, magnetostatic solution type has been used. With exciting the stator slots; flux lines, flux density distribution, flux vectors etc. have been obtained.

### 3.1. RMxprt

After entering necessary parameters into RMxprt, main motor configuration can be obtained. All necessary parameters have been given in '2.Design Parameters' section. Although some parameters have not been given in '2.Design Parameters' section the refecence speed has been assumed as the rated speed, slot liner has been taken as 0.3 mm, and limited fill factor has been taken as 0.75 .

With the parameters given above, our induction motor model has been obtained. This motor has been imported to Maxwell-2D, flux lines and flux densities of the machine has been examined. These results can be seen in '3.2.Maxwell 2D' section. It has been realized that the flux densities of the machine is too high, and saturation exists especially between the stator teeth. Because of this problem, machine parameters have to be modified. <br\>


To decrease the flux densities in the stator teeth, stator tooth width has been increased; also, the outer diameter of the stator has been increased.

With new parameters, optimized machine model and its windings can be seen in the following figures.

### 3.2. Maxwell 2D

In Maxwell 2D FEA software, magnetostatic solution has been used as solution type. The designed machine in RMxprt has been imported to Maxwell 2D the with exciting the stator slots in phase-A as 750 A positive, phase-B as 375 negative and phase-C as 375 negative; then flux lines, flux density distribution, and magnetic field distribution have been obtained

In the first iteration stator outer diameter and tooth width were inappropriate to obtain desired flux density distribution. 

```python
from IPython.display import Image
Image(filename='flux-first.png', width=800, height=800)
```

It can be observed that the flux density in stator teeth is around 2.2 T which is not acceptable because higher than 2 T, magnetic material goes into the saturation.

After some iterations, machine characteristic has been improved compared to the old one. In the following figures, results of magnetostatic analysis in Maxwell 2D can be seen.

_**Flux Lines**_
