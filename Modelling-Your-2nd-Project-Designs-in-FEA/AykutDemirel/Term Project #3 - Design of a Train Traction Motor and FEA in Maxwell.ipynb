**A. ANALYTICAL DESIGN**

First of all we need to assign values for the given parameters:

```python
Pmech = 1280000.0 # Watts
Vll = 1350.0 # Volts
polepair = 3.0
nrated = 1520.0 # rpm
torquerated = 7843.0 # Newton.meters
diameter_wheel = 1210.0 # meters
vmax = 38.89 # m/s
gear_ratio = 4.821
f = 78.0 # Hz
```

** Determination of main dimensions **

Althuogh it is hard to enqueue the steps of the motor design because of its being an iterative process, it is a good choice to start with determining the main machine dimensions according to standards and past experiences.

In the first step, we will set airgap diameter or and the axial length of the core. There is a parameter, called Cmech (machine constant), which indicates electrical and magnetic loading of the machine and can take values from the following interval:

```python
from IPython.display import Image 
Image(filename='Cmech.PNG')
```

In the above graph, Cmech values with respect to pole-power are given.

```python
pole_power = Pmech/(2*polepair)
print 'Pole Power = ',pole_power/1000.0,'kW'
```

From the graph, it is seen that we can start with Cmech = 240 kW.s/m^3.

```python
Cmech = 240000.0 # W.s/m^3
```

Cmech is related to machine dimensions with the expression: Cmech = Pmech / (D^2 x lp x nsynch)

Pmech is the mechanical power.

D is the stator inner diameter.

lp is equivalent axial length.

nsynch is the synchronous rotor speed in Hz.

```python
nsynch = f/polepair
```

To be able to find D and l value from the above equation, we need to know aspect ratio of the motor, which is X = l/D. Aspect ratio for a 6 pole motor is given by this general equation: 

X = (pi/2*polepair)*polepair^1/3

```python
X=(pi/(2*polepair))*(polepair**(1.0/3.0))
Dis = (Pmech/(Cmech*X*nsynch))**(1.0/3.0)
Dis = round(Dis*100.0)/100 # to get feasable numbers for dimensions
lp = Dis*X
lp = round(lp*100.0)/100.0

print 'Stator Inner Diameter = ', Dis*1000.0,'mm'
print 'Equivalent Axial Length = ', lp*1000.0,'mm'
```

Here, there is a mechanical constraint that we should consider. It is tip speed of the rotor. We need to calculate the tip speed of the rotor for the maximum rotor speed and confirm that it is sufficiently small. Rated speed of the train is 72 km/h and the maximum speed is given as 140 km/h. Then, the maximum rotor speed in rpm is given by:

```python
nmax=nrated*140.0/72.0 #rpm
```

Tip speed of the rotor for the maximum speed approxiamtely equals to:

```python
tipspeed = (nmax/60.0)*pi*Dis #m/s
print 'Tip speed = ', tipspeed,'m/s'
```

The resulting tip speed is approximately 100 m/s and it is in the acceptable.

Determining the inner stator diameter and the axial length of the motor, the next step will be assigning airgap value.

** Determination of the airgap **

The airgap value for a 6 pole machine is given by the following equation. Note that airgap value should be at least 0.2 because of mechanical constraints.

g = 0.18 + 0.006(Pmech)^0.4

```python
g = 0.18 + 0.006*((Pmech)**0.4) #in mm
g = round(g)
g = g/1000.0 # in m
print 'Airgap = ', g*1000.0,'mm'
```

** Winding selection **

Determining the outer stator diameter and the airgap of the motor, the next step is to assign number of stator slots and form a winding diagram.

Number of slots (Q) must be a multiple of number of phases and number of poles, which is 18 in our case. Slot pitch should be 7 mm at least due to manifacturing reasons.

```python
from IPython.display import Image 
Image(filename='slotpitch.PNG')
```

```python
innerstatorcircumferance = Dis*pi
tus = innerstatorcircumferance/54
print 'Slot pitch = ', tus*1000.0,'mm'
```

The slot pitch calculated is well in the range given in the above figure. Therefore number of stator slots can be assigned as 54.

```python
Qs = 54.0
qs = Qs/(polepair*2)/3.0 #number os slots per pole per phase
```

We can use double layer windings with 8/9 winding pitch. Then winding factor for the stator becomes:

```python
# kws = kps * kds  short pitch and distribution factors.
kps = sin(8.0*pi/18.0)
kds = sin(pi/6)/(qs*sin(pi/(6*qs)))
kws = kps * kds
print 'Stator winding factor = ', kws
```

From the below table, number of rotor slot can be chosen as 72 with 10 grad skew.

```python
Qr = 72.0
from IPython.display import Image 
Image(filename='rotorslotnumber.PNG')
```

** Airgap flux density **

We have already assigned a Cmech value which is realted to magnetic and electric loadings of the machine. For the airgap flux density, 0.8 T is a good starting point according to below table.

```python
Bairgap = 0.8 # Tesla
from IPython.display import Image 
Image(filename='magneticloading.PNG')
```

Cmech is related to magnetic and electric loadings by the followig equation:

Cmech = (pi^2) x kws x A x Bairgap / 2

,where A is the linear current density of the machine. We assigned Bairgap and we need to find A and confirm that it is in the range given in below table.

```python
from IPython.display import Image 
Image(filename='electricloading.PNG')
```

```python
A = 2.0*Cmech/ (pi*pi*kws*Bairgap)
print 'Linear current density = ', A/1000.0 , 'kA/m'
```

We see that the resulting electric loading is in the range given above and we can proceed with the design steps.

** No load flux and number of winding turns **

Since we have determined the main dimensions, winding schemes and the airgap flux density, we now need to determine number of coil turns in series in a phase.

Ns is given by the equation:

Ns = (2^0.5) x Em / (w x kws x lp x tp x ai x Bairgap)

* Em is EMF value.
* w is angular electrical speed.
* kws is winding factor.
* lp is equivalent length of the machine.
* tp is pole pitch.
* ai is a coefficient to find arithmetical avarage of the flux density of one pole..
* Bairgap is airgap flux density.

Em is approximately 0.98xUt where Ut is the terminal voltage.

tp can be founded by Dis x pi / polepair / 2

```python
Ut = Vll/((3.0)**0.5)
Em = 0.98*Ut
tp = Dis*pi/(polepair*2)

Ns = round((2.0**0.5)*Em/ (2*f*kws*lp*tp*2*Bairgap))

# number of conductors per slot should be an even integer because of double layer winding. So Ns should be rounded to an appropriate number.
Ns = 36.0
a = 1.0 # number of parallel branches.
zq = 2.0*a*3.0*Ns/Qs

print 'Number of coil turns per pole per phase = ', Ns
print 'Number of conductors per slot = ', zq
```

Since we changed Ns, we should calculate new airgap flux density.

```python
Bairgap = (2.0**0.5)*Em/ (2*f*kws*lp*tp*2*Ns)
print 'Airgap flux density = ', Bairgap, 'Tesla'
```

The new airgap flux density has a proper value.

** Determination of tooth width **

Determining the airgap flux density, we now should determine dimensions of the stator and rotor teeth. From the previously given magnetic loading tables, flux density of 1.7 T for stator and rotor teeth is good.

The stator teeth width, bds, is given by the equation:

bds = lp x tus x Bairgap / (kFe x l x Bstatorteeth) + 0.1 mm  where 

l and lp are real and effective length of the machine,

tus is slot pitch,

kFe is space factro of the iron,

Bairgap is airgap flux density,

Bstatorteeth is stator teeth flux density.

For simplicity, we can take lp/l and kFe as 1 and find an approximate width as follows:

```python
Bstatorteeth = 1.7 # Tesla
bds = tus * Bairgap/Bstatorteeth + 0.0001
print 'Stator tooth width = ', bds*1000.0, 'mm'
```

Tooth width should be a proper number for production tolerances. Therefore:

```python
bds = 0.014 # in meters
print 'Stator tooth width = ', bds*1000.0, 'mm'
```

By the same manner, rotor tooth width is:

```python
tur = pi*Dis / Qr
Brotorteeth = Bstatorteeth
bdr = tur * Bairgap/Brotorteeth + 0.0001
print 'Rotor tooth width = ', bdr*1000.0, 'mm'
```

Tooth width should be a proper number for production tolerances. Therefore:

```python
bdr = 0.011 # in meters
print 'Rotor tooth width = ', bdr*1000.0, 'mm'
```

** Determination of slot dimensions **

To be able to find stator and rotor slot dimensions, first we need to define how much area we need in the slots. Therefore, we need currents and current densities.

Stator current can be found by assuming 97% efficiency and 0.9 pawer factor as follows:

```python
Is = Pmech / (3*(Vll/(3.0**0.5))*0.97*0.9)
Ir = zq*Qs*Is*0.9/(4*Qr)
from IPython.display import Image 
Image(filename='currentdensity.PNG')
```

By looking at the above current density values, since our efficiency is high and our motor is a totally enclosed induction motor; we can choose a value for current densities close to the lower bound of the given allowable density range.

```python
J = 3.5 # A/mm^2
statorconductorarea = Is/(J*1000000) #m^2
rotorconductorarea =  Ir/(J*1000000) #m^2
print 'Stator conductor area = ', statorconductorarea*1000000, 'mm^2'
print 'Rotor conductor area = ', rotorconductorarea*1000000, 'mm^2'
```

For the stator windings we use 27 parallel AWG9 size cables to make stator coils easy to bend and to fabricate and to eliminate skin effect at 78 Hz. AWG9 size cable has 6,62 mm^2 crosssectional area. Randomly wounded coils also introduces a 0.7 space factor coefficient to the slot area and stator slot area finally becomes:

```python
Astatorslot = 27.0*0.00000662*zq/0.7
print 'Stator slot area = ', Astatorslot*1000000, 'mm^2'
```

Since rotor bars are solid copper, there are no conductor selection or space factor:

```python
Arotorslot = rotorconductorarea
print 'Rotor slot area = ', Arotorslot*1000000, 'mm^2'
```

Based on the above values, stator and rotor slot openings are visualized in Motor Analysis Software. The approximate slot widths and slot depths of stator and rotor are calculated by assuming the slot openings are rectangular and their thicknesses are not changing in outward direction from the center. Although this is a rough approxiamtion, it is sufficient for us to get machine characteristics. Detailed geometric calculations can be done in the production process, later.

```python
statorslotwidth = tus - bds
rotorslotwidth = tur - bdr
statorslotdepth = Astatorslot/statorslotwidth
rotorslotdepth = Arotorslot/rotorslotwidth
print 'Stator slot depth = ', statorslotdepth*1000, 'mm'
print 'Stator slot width = ', statorslotwidth*1000, 'mm'
print 'Rotor slot depth = ', rotorslotdepth*1000, 'mm'
print 'Rotor slot width = ', rotorslotwidth*1000, 'mm'
```

** Determination of stator outer and rotor inner diameters **

We know the airgap flux density, we know the teeth flux densities and the remaining part is the yoke flux densities and rotor inner and stator outer diameters. From an earlier figure, we can say that flux density of 1.5 T for stator yoke and 1.2 T for rotor yoke is well. Without doing complex calculations, those diameters can be calculated using the ratio of airgap flux density and yoke flux densities as follows:

```python
Bstatoryoke = 1.0 #Tesla
Brotoryoke = 0.8 #Tesla
Fluxperpole = Dis*pi/(polepair*2.0)*lp

# Dos will be stator outer diameter

# Bstatoryoke = (Fluxperpole/2)/(lp x (Dos-Dis-statorslotdepth))

#Therefore;

Dos = ((Fluxperpole/2.0)/(lp*Bstatoryoke))+Dis+statorslotdepth

#in the same way the rotor inner diameter becomes:

Dir = Dis-rotorslotdepth-((Fluxperpole/2.0)/(lp*Brotoryoke))

print 'Outer stator diameter = ', Dos*1000.0, 'mm'
print 'Inner rotor diameter = ', Dir*1000.0, 'mm'
```

Completing the analytical design of the machine, now we can proceed with the FEA in MAXWELL.

** C. CONCLUSIONS **


In the conclusion part, general comments about machine design and EE564 course will be given based on some evaluations from this design.

* Basic machine dimensions

As a first step of the machine design, rough esitmations about the overall dimensions of the machine can be made given the power rating and the pole number of a machine. 

Before this course and the 3 projects assigned, we hardly visualize in our minds how big a 1 MW motor is but now we can estimate the dimesions using proper magnetic and electrical loading values. Cmech value is a representative parameter that determines these loadings. Based on this, the output power or the torque of a machine can be increased by increasing the Cmech value or increasing the machine dimensions.

* Winding selection

Based on the experience gained from the course, winding is the actuating part of the machine. Distribution and pitch factor of the selected winding diagram affect the design performance significantly. Proper pitch factors can be selected to eliminate some harmonics in the design and increase the overall performance.

* Detailed dimensions

The machine dimensions in detail are decided not only according to analytical calculation but also considering the ease of manufacturing. We may design a perfect machine but if it is not relizable in real life, it is nothing at all. The most important thing learned from this course is the real life considerations of a machine design besides the thoretical calculations.

* Material selection

Material selection is the other important step of the machine design that directly determines the magnetic and electric loadings and therefore the power rating of the machine. Machine materials should be chosen for calculated current and flux densities appropriately. The winding cables, their current carrying capabilities, insulation levels should be OK for the specific design. Moreover, the magnetic properties of the rotor and stator core materials should be properly selected in terms of magnetic loading. In this manner, operation frequency is a key feature and it directly plays with the core loss of the selected materials.

* Thermal considerations

Most of the times, the limiting factor of an electrical machine is the thermal performance. The loss inside the machine should be disposed outside of the machine properly for machine to continue its operation. As the temperature increases in the machine, the machine parameters change and the machine may not perform as designed anymore. Even more catastrophic effects may occure and the machine may be mechanically defected due to excess heat dissipated inside the machine.

In thermal design of the machine, working conditions become important. A machine should be designed according to its purpose and working condition to meet the requirements with a minimum material consumption and engineering.

* Simulation programmes

Since design of an electrical machine is a difficult process, it is not possible to manifacture a prototype over and over again by making small differences and see the results. Here the most useful tools that can be used becomes simulation programmes based on finite element analysis. One of such programmes is Ansoft - Maxwell Software and it is used in this project. 

Maxwell is a useful tool by which we can see the performance of the machine and predict any problems before manifacturing the machine.

In summary, machine design is an iterative process in which each step may affect the previous ones. There is not the best machine but there is the optimum machine for a specific application in terms of design. There are not all application specific machines but there are standart machine classes to be selected in terms of manifacturing.
