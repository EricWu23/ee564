# EE564 THIRD PROJECT

The third project of the semester is the modelling the induction motor (IM) design which is analytically calculated in the second project.

Anysys Maxwell is used as FEA method to model the IM machine in detail. RMxpert and Maxwell 2D analysis tools are utilized.

Detailed information on the second project can be obtained [here](https://github.com/odtu/ee564/tree/master/TESLA/Nadir)

IM is designed for Tesla Model S RWD 85 model with the following specs:

* Max. Power: 360 hp (270 kW)
* Max. Torque: 441 Nm
* Top Speed: 225 km/h

The analysis of IM is achieved in RMxpert tool firstly. Then, the model is transferred to Maxwell 2D environment.

In Maxwell 2D, the modelled machine is analysed in both **transient** and **magnetostatic** solution types.

**Attention:**

The reults obtained in both RMxpert and Maxwell 2D are not found to be satisfactory due to high flux densities in the stator tooth and stator yoke (back core).

Therefore, the designed machine is updated in terms of stator slot dimensions. The tooth width is increased to 6mm and the slot height is decreased to 36mm.

As a result of decreasing the slot area, the slot fill factor increased to **82%** which can be regarded as a high fill factor but the machine is **_liquid cooled_** hence the cooling will not be aproblem for such a high current density in stator slots.
