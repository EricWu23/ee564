# EE564 THIRD PROJECT

The third project of the semester is the modelling the induction motor (IM) design which is analytically calculated in the second project.

Anysys Maxwell is used as FEA method to model the IM machine in detail. RMxprt and Maxwell 2D analysis tools are utilized.

Detailed information on the second project can be obtained [here](https://github.com/odtu/ee564/tree/master/TESLA/Nadir)

IM is designed for Tesla Model S RWD 85 model with the following specs:

* Max. Power: 360 hp (270 kW)
* Max. Torque: 441 Nm
* Top Speed: 225 km/h

The analysis of IM is achieved in RMxprt tool firstly. Then, the model is transferred to Maxwell 2D environment.

In Maxwell 2D, the modelled machine is analysed in both **transient** and **magnetostatic** solution types.

**Attention:**

The reults obtained in both RMxprt and Maxwell 2D are not found to be satisfactory due to high flux densities in the stator tooth and stator yoke (back core).

Therefore, the designed machine is updated in terms of stator slot dimensions. The tooth width is increased to 6mm and the slot height is decreased to 36mm.

As a result of decreasing the slot area, the slot fill factor increased to **82%** which can be regarded as a high fill factor but the machine is **_liquid cooled_** hence the cooling will not be aproblem for such a high current density in stator slots.

## 2. Maxwell 2D Magnetostatic Solution

Magnetostatic solution requires excitation of stator windings.

The fraction is selected as 1 hence the whole machine is simulated. The rated current is approximated as **600Arms** in Phase A. Phases B and C therefore has a rated current of **300A in the negative direction**.

In this way, all the windings are excited and the machine is simulated. The following results are obtained:

* Mesh Plot
* Air Gap Flux Density of the quarter machine
* Air Gap Magnetic Field of the quarter machine
* B Field
* Flux Linkages
* J Field

```python
Image(filename='images\Magnetostatic\Magnetostatic Mesh Plot.png', width=600)
```

The machine is **4 pole** and symmetric. Hence the air gap flux density of a quarter machine is observed using a polyline in Maxwell 2D.

The flux density barely exceeds **1 Tesla**.

```python
Image(filename='images\Magnetostatic\Air Gap Flux Density quarter machine.png')
```

```python
Image(filename='images\Magnetostatic\Air Gap Magnetic Field quarter machine.png')
```

The flux density on the whole machine is coherent with the RMxprt data. The maximum density occurs in the stator shoes as expected. A way to decrease the flux density is to increase the diameter of the motor hence the tooth width increases.

```python
Image(filename='images\Magnetostatic\Magnetostatic B Field.png', width=600)
```

Flux lines reflects that the motor is a 4 pole one. The flux lines are tracing the back core. If the flux lines trace stator slots (inside), then the back core shall be increased to provide a suitable path for the flux.

```python
Image(filename='images\Magnetostatic\Magnetostatic Flux Linkages.png', width=600)
```

```python
Image(filename='images\Magnetostatic\Magnetostatic J Field.png', width=600)
```

## 3. Maxwell 2D Transient Solution

Different from the magnetostatic solution, predefined phase voltages are applied to windings in transient solution. Moreover, transient solution requires eddy effects to be applied.

The fraction is selected as 1 hence the whole machine is simulated. The phase voltage is then **231*sin(2*pi*171*time)** in Phase A. Phases B and C therefore has 120 degrees shifted voltages.

In this way, all the windings are excited and the machine is simulated during **200ms**. The following results are obtained:

* Input Voltages
* Induced Voltages
* Winding Currents
* Flux Linkages
* Moving Torque
* Mesh Plot
* A Field
* B Field
* H Field

## 3. Conclusion

* IM design of 2nd project is implemented in Ansys Maxwell.
* The analytical results are coherent with the FEA results.
* A difference in flux densities in the stator teeth occurs. This is solved by increasing tooth width and decreasing slot height (increasing back core).
* Rated torque can be obtained up to rated speed using vector controlled inverter. Beyond rated speed, constant power region is reached.
* A top to bottom design procedure of electrical machines is learned throughout this course. The design starts with stating design parameters, continues with experimentally calculated constants and analytical calculations. At the end or at an intermeddiate step, design has to be iterated in order to optimize the machine parameters. Therefore, the whole procedure needs to be reproducible. IPython is a useful tool to obtain a reproducible design procedure. Before production, the machine needs to be verified by means of a FEA tool such as Flux, Maxwell etc. Detailed simulation results are obtained in FEA.
* Maxwell offers to use optimization of desired machine parameters in 2D analysis; however, it requires detailed information on machine design and may not be needed for the time being.
